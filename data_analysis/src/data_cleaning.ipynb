# Limpeza e padronização

```python
data = data.copy()
```

```python
print(data.columns)
```

```python
data.columns = data.columns.str.replace(' ','')
print(data.columns)
```

```python
data.isnull().sum()
```

```python
data = data.replace('-', '', regex=True)
data = data.replace(',','', regex=True)
data = data.replace('%','', regex=True)
data = data.replace('', np.nan)

data = data.dropna()

data.isnull().sum()
```

```python
data.dtypes
```

```python
data = data.astype("float64")
```

```python
data.dtypes
```

# Gerando nomes para pacientes do dataset

```python
faker = Faker()

names = []

for i, row in data.iterrows():
    if 'sex' == 0:
        names.append(faker.name_male())
    else:
        names.append(faker.name_female())


data.to_csv('../data/processed/cleaned_data.csv', index=False)
```
