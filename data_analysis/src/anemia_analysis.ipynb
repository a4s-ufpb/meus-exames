# Discretização dos valores das taxas

Hemoglobina

```python
condition_hgb = [
    # Homens (Sex == 0.0)
    (filtered_data["Sex"] == 0.0) & (filtered_data["HGB"] < 12),
    (filtered_data["Sex"] == 0.0) & (filtered_data["HGB"] > 15),

    # Mulheres (Sex != 0)
    (filtered_data["Sex"] != 0.0) & (filtered_data["HGB"] < 11),
    (filtered_data["Sex"] != 0.0) & (filtered_data["HGB"] > 14)
]

values_hgb = [
    0, # Homens com HGB < 12
    2, # Homens com HGB > 15
    2, # Mulheres com HGB < 11
    0 # Mulheres com HGB > 14
]

filtered_data["HGB_Class"] = np.select(condition_hgb, values_hgb, default=1)
```

Hematócrito

```python
condition_pcv = [
    # Homens (Sex == 0.0)
    (filtered_data["Sex"] == 0.0) & (filtered_data["PCV"] < 38.3),
    (filtered_data["Sex"] == 0.0) & (filtered_data["PCV"] > 48.6),

    # Mulheres (Sex != 0)
    (filtered_data["Sex"] != 0.0) & (filtered_data["PCV"] < 35.5),
    (filtered_data["Sex"] != 0.0) & (filtered_data["PCV"] > 44.9)
]

values_pcv = [
    0, # Homens com PCV < 38.6
    2, # Homens com PCV > 48.6
    2, # Mulheres com PCV < 35.5
    0 # Mulheres com PCV > 44.9
]

filtered_data["PCV_class"] = np.select(condition_pcv, values_pcv, default=1)
```

Concentração de hemoglobina corpuscular média

```python
condition_mchc = [
    filtered_data["MCHC"] > 38,
    filtered_data["MCHC"] < 30
]

values_mchc = [0,2]

filtered_data["MCHC_class"] = np.select(condition_mchc, values_mchc, default=1)
```

Hemoglobina corpuscular médio

```python
condition_mch = [
    filtered_data["MCH"] > 35,
    filtered_data["MCH"] < 25
]

values_mch = [0,2]

filtered_data["MCH_class"] = np.select(condition_mch, values_mch, default=1)
```

Amplitude de distribuição das hemácias

```python
condition_rdw = [
    filtered_data["RDW"] > 16,
    filtered_data["RDW"] < 12.5
]

values_rdw = [0,2]

filtered_data["RDW_class"] = np.select(condition_rdw, values_rdw, default=1)
```

Plaquetas

```python
condition_plt = [
    filtered_data["PLT/mm3"] > 450,
    filtered_data["PLT/mm3"] < 140
]

values_plt = [0,2]

filtered_data["PLT/mm3"] = np.select(condition_plt, values_plt, default=1)
```

Hemácias

```python
condition_rbc = [
    filtered_data["RBC"] > 5.8,
    filtered_data["RBC"] < 4
]

values_rbc = [0,2]

filtered_data["RBC_class"] = np.select(condition_rbc, values_rbc, default=1)
```

Volume corpuscular médio

```python
condition_mcv = [
    filtered_data["MCV"] > 115,
    filtered_data["MCV"] < 75
]

values_mcv = [0,2]

filtered_data["MCV_class"] = np.select(condition_mcv, values_mcv, default=1)
```

Contagem total de leucócitos

```python
condition_tlc = [
    filtered_data["TLC"] > 115,
    filtered_data["TLC"] < 75
]

values_tlc = [0,2]

filtered_data["TLC_class"] = np.select(condition_tlc, values_tlc, default=1)
```

# Discretização da idade em faixas etárias.

```python
filtered_data["Age_group"] = pd.cut(
    filtered_data["Age"],
    bins=[-float("inf"), 32,48,64, float("inf")],
    labels = [0,1,2,3]
).astype(int)
```

Faixas etárias:
* 0: ≤ 32
* 1: 33–48
* 2: 49–64
* 3: ≥ 65
